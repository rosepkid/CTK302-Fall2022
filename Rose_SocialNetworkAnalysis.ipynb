```python
import networkx as nx

Andor = nx.Graph() 
Andor.add_nodes_from(['Bix', 'Cassian', 'Mon', 'Maerva', 'Leia', 'Luthen', 'Syril', 'Saw', 'Dedra', 'Vel', 'Cinta'])

```

```python
#create a network for who spoke to whom
Spoke=nx.DiGraph()
Spoke.add_edges_from([('Bix','Cassian'),('Luthen','Bix'),('Luthen','Vel'),('Vel','Cinta'),('Cassian','Maerva'),('Luthen','Saw')], label='spoke to')
spokeDict=nx.get_edge_attributes(Spoke,'label')
```

```python
Allies=nx.Graph()
Allies.add_edges_from([('Bix','Cassian'),('Cassian','Luthen'),('Vel','Cinta'),('Mon','Luthen'),('Syril','Dedra'),('Cassian','Maerva')],label='are allies')
alliesDict=nx.get_edge_attributes(Allies,'label')
```

```python
import matplotlib.pyplot as plt

#setting up the drawing area and layout
plt.figure(1,figsize=(6,6))
pos=nx.circular_layout(Andor)

#show the nodes
nx.draw(Andor, pos, node_color='grey', node_size=5, with_labels=True)

#draw the speaking edges
nx.draw_networkx_edges(Spoke, pos, edge_color='red')

#ading purple Spoke labels to the graph
nx.draw_networkx_edge_labels(Spoke,pos,edge_labels=spokeDict, label_pos=0.2, font_color='purple')

#draw the allies edges
nx.draw_networkx_edges(Allies, pos, edge_color='blue')

#ading purple Allies labels to the graph
nx.draw_networkx_edge_labels(Allies,pos,edge_labels=alliesDict, label_pos=0.5, font_color='green')
plt.show()
```
